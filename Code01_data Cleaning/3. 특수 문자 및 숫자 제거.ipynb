#### 숫자 및 특수문자 제거

```python
import re

def remove_special_chars(cell_value):
    if isinstance(cell_value, str):
        pattern = re.compile(r'[^a-zA-Zㄱ-힗]')
        return pattern.sub('', cell_value)
    else:
        return cell_value
```

```python
df_agoda = df_agoda.applymap(remove_special_chars)

df_agoda
```

#### 5글자 이하 제거



```python
df_agoda = df_agoda[df_agoda['content'].str.len()>5]

df_agoda
```
