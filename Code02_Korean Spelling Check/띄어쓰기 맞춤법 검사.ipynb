### 데이터 불러오기

```python
import pandas as pd

df_agoda = pd.read_csv('/content/drive/MyDrive/리뷰데이터 분석/데이터셋/데이터함/특수문자 및 숫자 제거(5글자이상)/df_agoda.csv', encoding='utf-8')
df_airbnb = pd.read_csv('/content/drive/MyDrive/리뷰데이터 분석/데이터셋/데이터함/특수문자 및 숫자 제거(5글자이상)/df_airbnb.csv', encoding='utf-8')
df_booking = pd.read_csv('/content/drive/MyDrive/리뷰데이터 분석/데이터셋/데이터함/특수문자 및 숫자 제거(5글자이상)/df_booking.csv', encoding='utf-8')
df_dailyhotel = pd.read_csv('/content/drive/MyDrive/리뷰데이터 분석/데이터셋/데이터함/특수문자 및 숫자 제거(5글자이상)/df_dailyhotel.csv', encoding='utf-8')
df_hotelsdotcom = pd.read_csv('/content/drive/MyDrive/리뷰데이터 분석/데이터셋/데이터함/특수문자 및 숫자 제거(5글자이상)/df_hotelsdotcom.csv', encoding='utf-8')
df_myrealtrip = pd.read_csv('/content/drive/MyDrive/리뷰데이터 분석/데이터셋/데이터함/특수문자 및 숫자 제거(5글자이상)/df_myrealtrip.csv', encoding='utf-8')
df_yanol = pd.read_csv('/content/drive/MyDrive/리뷰데이터 분석/데이터셋/데이터함/특수문자 및 숫자 제거(5글자이상)/df_yanol.csv', encoding='utf-8')
df_yeogi = pd.read_csv('/content/drive/MyDrive/리뷰데이터 분석/데이터셋/데이터함/특수문자 및 숫자 제거(5글자이상)/df_yeogi.csv', encoding='utf-8')
```

```python
all_OTA = pd.concat([df_agoda,df_airbnb,df_booking,df_dailyhotel,df_hotelsdotcom,df_myrealtrip,df_yanol,df_yeogi],ignore_index=True)

all_OTA
```

```python
all_OTA = all_OTA.drop(columns = ['Unnamed: 0','date','date.1','score','at'])
```

### 띄어쓰기 진행

```python
pip install git+https://github.com/haven-jeon/PyKoSpacing.git
```

```python
from pykospacing import Spacing
from tqdm import tqdm
import time

spacing = Spacing()

for i in tqdm(range(len(all_OTA.content))):
    try:
        all_OTA.content[i] = spacing(all_OTA.content[i]) 

    except Exception as e:
        print(f"Error occurred at index {i}: {e}")
        continue

print(all_OTA)
```

### 맞춤법 검사 진행

```python
!pip install git+https://github.com/ssut/py-hanspell.git
```

```python
from hanspell import spell_checker 
from tqdm import tqdm
import time

for i in tqdm(range(len(all_OTA.content))):
    try:
        spelled_sent = spell_checker.check(all_OTA.content[i]) 
        all_OTA.content[i] = spelled_sent.checked 

    except Exception as e:
        print(f"Error occurred at index {i}: {e}")
        continue

print(all_OTA)
```
