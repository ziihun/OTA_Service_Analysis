```python
from pykospacing import Spacing
from hanspell import spell_checker 
from tqdm import tqdm
import time
import numpy as np

spacing = Spacing()
```

```python
def process_dataframe(df):
    for i in tqdm(range(len(df.content))):
        try:
            df.content[i] = spacing(df.content[i])
            spelled_sent = spell_checker.check(df.content[i]) 
            df.content[i] = spelled_sent.checked 

        except Exception as e:
            print(f"Error occurred at index {i}: {e}")
            continue
    
    df['sentiment'] = np.where(df['score'] >= 4, 1, 0)
    
    return df
```

```python
process_dataframe(df_agoda)
```

```python
process_dataframe(df_airbnb)
```

```python
process_dataframe(df_booking)
```
