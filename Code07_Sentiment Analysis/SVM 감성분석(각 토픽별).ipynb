```python
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer

def predict_sentiment(texts, model, vocabulary):
    okt = Okt()
    
    def preprocess(text):
        words = okt.morphs(text, stem=True)
        words = [word for word in words if word not in stopwords]
        return ' '.join(words)
    
    def calculate_tfidf(text, vocabulary):
        tfidf = TfidfVectorizer(vocabulary=vocabulary)
        tfidf.fit(text)
        return tfidf.transform(text)
    
    def predict_sentiment_for_text(text, model, vocabulary):
        preprocessed_text = preprocess(text)
        tfidf = calculate_tfidf([preprocessed_text], vocabulary)
        score = model.predict(tfidf)[0]
        return (score * 100) - 50
    
    scores = [predict_sentiment_for_text(text, model, vocabulary) for text in texts]
    return sum(scores) / len(scores) # 입력 text의 점수 평균값 리턴
```

### 함수 생성

```python
import numpy as np
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer

def sentiment_analysis(df, model, vocabulary, stopwords):
    okt = Okt()
    
    def preprocess(text):
        words = okt.morphs(text, stem=True)
        words = [word for word in words if word not in stopwords]
        return ' '.join(words)
    
    def calculate_tfidf(text, vocabulary):
        tfidf = TfidfVectorizer(vocabulary=vocabulary)
        tfidf.fit(text)
        return tfidf.transform(text)
    
    def predict_sentiment_for_text(text, model, vocabulary):
        preprocessed_text = preprocess(text)
        tfidf = calculate_tfidf([preprocessed_text], vocabulary)
        score = model.predict(tfidf)[0]
        return (score * 100) - 50
    
    # 열 'Dominant_Topic', 'text' 제거
    df = df.drop(columns=['Dominant_Topic', 'text'])
    
    # 빈값 제거
    df = df.dropna(how='any')
    
    # 한국어 아니면 제거
    df['content'] = df['content'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
    df['content'] = df['content'].str.replace('^ +', "")
    df['content'].replace('', np.nan, inplace=True)
    df = df.dropna(how='any')
    
    # okt 형태소 분석 및 stopwords 제거
    df['content'] = df['content'].apply(lambda x: preprocess(x))
    
    # TF-idf 계산 후 SVM 모델에 입력하여 -50~50사이 점수 추출
    scores = [predict_sentiment_for_text(text, model, vocabulary) for text in df['content']]
    
    # 평균 점수 return
    avg_score = sum(scores) / len(scores)
    return avg_score
```
